```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
train = pd.read_csv('Htrain.csv')
train.head()
```

```python
test = pd.read_csv('Htest.csv')
test.head()
```

```python
human = pd.concat([train,test],axis=0)
```

```python
human
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.cross_validation import train_test_split
```

```python
numeric = human._get_numeric_data()
category = human.drop(numeric.columns,axis = 1)
```

```python
category.head()
```

```python
train_dx = human.drop(["Activity"],axis = 1)
train_dy = human.Activity
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
        train_dx,
        train_dy,
        test_size=0.20,
        random_state=123)
```

```python
radm_clf = RandomForestClassifier(oob_score=True,n_estimators=100 )
radm_clf.fit( X_train, Y_train )
```

```python
indices = np.argsort(radm_clf.feature_importances_)[::-1]
feature_rank = pd.DataFrame( columns = ['rank', 'feature', 'importance'] )
for f in range(X_train.shape[1]):
    feature_rank.loc[f] = [f+1,
                         X_train.columns[indices[f]],
                         radm_clf.feature_importances_[indices[f]]]
f, ax = plt.subplots(figsize=(10,100))
sns.barplot( y = 'feature', x = 'importance', data = feature_rank, color = 'Yellow')
plt.show()
```

```python
ff = feature_rank.head(100)
ff
```

```python
for i in category:
    category[i] = category[i].astype('category')
```

```python
for i in category:
    category[i] = category[i].cat.codes
```

```python
category.Activity.value_counts()
```

```python
df = pd.concat([numeric,category],axis=1)
```

```python
df.head()
```

```python
from sklearn.decomposition import PCA
```

```python
train_d1= df.drop(['Activity'],axis = 1)
y = df.Activity
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(train_d1)                
t_train = scaler.transform(train_d1)
```

```python
pca_hp = PCA(0.99)
x_fit = pca_hp.fit_transform(t_train)
print (t_train.shape, x_fit.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_fit,y)
```

```python
X_train , X_test, Y_train, Y_test = train_test_split(
        x_fit,
        y,
        test_size=0.20,
        random_state=123)
```

```python
y_pred = log.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(Y_test, y_pred)
```

```python
cm = metrics.confusion_matrix(Y_test,
                              y_pred, [0,1,2,3,4,5])
sns.heatmap(cm, annot=True,  fmt='.2f', xticklabels = ["0","1","2","3","4" "5"] , yticklabels =  ["0","1","2","3","4" "5"] )
plt.ylabel('True label')
plt.xlabel('Predicted label')
```

```python
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier, export_graphviz, export 
from sklearn.grid_search import GridSearchCV
```

```python
depth_list = list(range(1,20))
for depth in depth_list:
    dt_obj = DecisionTreeClassifier(max_depth=depth)
    dt_obj.fit(X_train, Y_train)
    print ('depth:', depth, 'Accuracy:', metrics.accuracy_score(Y_test, dt_obj.predict(X_test)))
```

```python
param_grid = {'max_depth': np.arange(3,20)}
tree = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=10)
tree.fit(X_train, Y_train)
```

```python
tree.best_params_
```

```python
tree.best_score_
```

```python
tree_final = DecisionTreeClassifier(max_depth=8)
tree_final.fit(X_train, Y_train)
```

```python
tree_test_pred = pd.DataFrame({'actual': Y_test, 'predicted': tree_final.predict(X_test)})
```

```python
metrics.accuracy_score(Y_test, tree_test_pred.predicted)
```
