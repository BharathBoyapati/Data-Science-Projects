```python
titanic = pd.read_csv('titanictrain.csv')
titanic
```

```python
titanic = titanic.drop(['PassengerId','Cabin'], axis = 1)
```

```python
titanic.columns
```

```python
num_col = titanic._get_numeric_data()
num_col.columns
```

```python
titanic.Survived = pd.Categorical(titanic.Survived)
titanic.Pclass = pd.Categorical(titanic.Pclass)
titanic.SibSp = pd.Categorical(titanic.SibSp)
titanic.Parch = pd.Categorical(titanic.Parch)
```

```python
num_col = titanic._get_numeric_data()
num_col.columns
```

```python
def var_summary(x):
    return pd.Series([x.count(), x.isnull().sum(), x.sum(), x.mean(), x.median(),  x.std(), x.var(), x.min(), x.quantile(0.01), x.quantile(0.02),x.quantile(0.10),x.quantile(0.25),x.quantile(0.50),x.quantile(0.75), x.quantile(0.90),x.quantile(0.98), x.quantile(0.99),x.max()], 
                  index=['N', 'NMISS', 'SUM', 'MEAN','MEDIAN', 'STD', 'VAR', 'MIN', 'P1' , 'P2' ,'P10' ,'P25' ,'P50' ,'P75' ,'P90' ,'P98' ,'P99' ,'MAX'])

num_col.apply(lambda x: var_summary(x)).T
```

```python
num_col['Age'] = num_col['Age'].clip_upper(num_col['Age'].quantile(0.99))
num_col['Fare'] = num_col['Fare'].clip_upper(num_col['Fare'].quantile(0.99))
```

```python
num_col['Age'] = num_col['Age'].fillna(num_col['Age'].mean())
```
